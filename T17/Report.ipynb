```python
import pandas as pd
```

```python
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity. 
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`For a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)). 




```python
# Compares the average income based on ethnicity
df.groupby(['Ethnicity'])['Income'].mean()
```

```python
# Returns information on who have higher balance, married or single people
if df[df.Married == 'Yes'].loc[:,'Balance'].mean() > df[df.Married == 'No'].loc[:,'Balance'].mean():
    print('On average married couples have higher balance, in sum of', df[df.Married == 'Yes'].loc[:,'Balance'].mean(),'.')
else:
    print('On average single people have a higher balance, in sum of', df[df.Married == 'No'].loc[:,'Balance'].mean(),'.')
```

```python
# Returns highest income in the dataset
print('The highest income in our dataset is',df.loc[:,'Income'].max(),'.')
```

```python
# Returns lowest income in the dataset
print('The lowest income in our dataset is',df.loc[:,'Income'].min(),'.')
```

```python
# Returns amount of cards recorded in the dataset
print('We have recorded', df.loc[:, 'Cards'].sum(), 'cards in our dataset.')
```

```python
# Returns amount of females and males in the dataset
print('Males vs females - amount of entries by:\n ', df['Gender'].value_counts())
```
