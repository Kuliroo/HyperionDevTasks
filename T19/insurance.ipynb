```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
our_data = pd.read_csv("insurance.csv")
our_data.head()
```

```python
# Columns used for the model
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values
# Creating scatter plot
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charge')
plt.show()
```

```python
#Creating a Linear Regression model and fitting it to the data
charges_model = LinearRegression()
charges_model.fit(x,y)
```

```python
# Predictions model with the best-fit line
y_pred = charges_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charge')
plt.show()
```

```python
# Additional check from example file - Predict an unknown value
unk_x = [[99]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charge')
plt.show()

print("A person aged:", str(unk_x).strip("[]"), ", will have to pay:", str(charges_model.predict(unk_x)).strip("[]"))
```
